## The concepts behind
This notebook will introduce the concept behind the computational ToM by Devaine et al. (2017). The computational ToM is a model which seeks to approximate the model human uses a game theory context.



 It uses variational Bayes Laplace approximation to estimate opponent parameter values, sicne it is a way of approximating Bayes-optimal updating, while being computationally feasible enough so that it is a realistic model of the Bayesian brain, as opposed to traditional sampling methods (Friston et al., 2006a; Daunizeau, 2017b).
